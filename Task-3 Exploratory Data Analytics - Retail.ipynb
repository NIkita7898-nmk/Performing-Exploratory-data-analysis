### Import and Read Data

```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
#reading data
data=pd.read_csv("SampleSuperstore.csv")
data.head(10)
```

```python
#information of data
data.info()
```

We can see in the data there is no null value.

## Data Visualization

```python
sns.pairplot(data)
```

```python
fig=axes = plt.subplots(1,1,figsize=(12,7))
sns.heatmap(data.corr(),annot=True)
plt.show()
```

```python
fig,axes=plt.subplots(1,2,figsize=(14,5))
fig.suptitle("Total profit VS Sales")
sns.barplot(data=data.groupby('A')['Sales','Profit'].agg(sum),x='Sales',y='Profit',ax=axes[1])
data.groupby('A')['Sales','Profit'].agg(sum).plot(kind='bar',ax=axes[0])
plt.xticks(rotation=90)
plt.show()
```

```python
fig,axes=plt.subplots(1,2,figsize=(14,5))
data.groupby('A')['Profit','Discount'].agg(sum).plot(kind='bar',ax=axes[0]).set_title('Quality & Profit Relation based on A')
data.groupby('A')['Profit','Quantity'].agg(sum).plot(kind='bar',ax=axes[1]).set_title('Discount & Profit Relation based on A')
plt.xticks(rotation=90)
plt.show()
```

```python
fig,ax=plt.subplots(1,1,figsize=(9,6))
sns.countplot(data['Quantity'],hue=data['Region'])
plt.show()
```

```python
fig,axes=plt.subplots(2,2,figsize=(11,8))
fig.suptitle("Sales with different shipping modes and Segments",fontsize=16)
sns.barplot(data['Ship Mode'],data['Sales'],ax=axes[0,0])
sns.lineplot(data['Ship Mode'],data['Sales'],ax=axes[0,1])
sns.barplot(data['Segment'],data['Sales'],ax=axes[1,0])
sns.lineplot(data['Segment'],data['Sales'],ax=axes[1,1])
plt.show()
```

#### The maximum quantity  of products in demand was in range 2 - 4.

#### The features profit and discounts are highly related.
